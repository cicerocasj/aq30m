```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

<h3>Leitura dos dados: Verifica 1 e Verifica 3</h3>

```python
df = pd.read_csv('conhecimento.csv')

df1 = df[df.verifica==1].copy().reset_index(drop=True)
df3_full = df[df.verifica==3].copy().reset_index(drop=True)

ids_3 = np.random.choice(df3_full.id, len(df1), replace=False)
df3 = df[df.id.isin(ids_3)].copy().reset_index(drop=True)
```

```python
cols = ['ndvi', 'nbrl', 'dif_ndvi', 'dif_dnbrl', 'medianb2', 'medianb3',
       'medianb4', 'medianb5', 'medianb6', 'medianb7']
```

<h3>Comparação entre verifica 1 e 3 para cada atributo</h3>

```python
for col in cols:
    fig, ax = plt.subplots()
    ax.set_title(col)
    df1[col].cumsum().plot(ax=ax, legend=False, c='r', figsize=(15,8), label='verifica 1')
    df3[col].cumsum().plot(ax=ax, legend=False, c='b', figsize=(15,8), label='verifica 3')
    ax.legend(loc='upper left')
```

<h3>Conclusão</h3>
<pre>
- O atributo <b>nbrl</b> é o único que tem direção diferente entre verifica 1 e 3
- O atributo medianb6 apresenta a maior osilação nos valores
- Os demais atributos mostram que existe uma pequena diferença entre o somatórios.
- Apesar de ter uma pequena diferença das linhas, ambas vão para a mesma direção.
</pre>
